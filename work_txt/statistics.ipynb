```python
import os
import sys
import inspect
from collections import Counter
from glob import glob
from pymystem3 import Mystem
```

```python
# –ü–æ–ª—É—á–µ–Ω–∏–µ —Å–ø–∏—Å–∫–∞ —Å—Ç–æ–ø —Å–ª–æ–≤
stop_words = []
with open('/home/al/Projects_My/hakaton_ozon_opg/dataset/stop_word_base.txt', 'r') as f:
    for i in f:
        stop_words.append(i.strip())
stop_words
```

```python
def get_dir(NAME_START: str) -> list:
    '''–ü–æ–ª—É—á–µ–Ω–∏–µ —Å–ø–∏—Å–∫–∞ –¥–∏—Ä–µ–∫—Ç–æ—Ä–∏–π –Ω–∞–±–æ—Ä–æ–≤
    '''
    dd = [ f.path for f in os.scandir(NAME_START) if f.is_dir() ]
    return dd
```

```python
def get_array_text(sum_arr: dict, key: str, nf: str, m):
    """ –ü–æ–ª—É—á–µ–Ω–∏–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ —Å–ª–æ–≤ –ø–æ –≤–∏–¥–∞–º —Ç–æ–≤–∞—Ä–æ–≤

    Args:
        sum_arr (dict): _description_
        key (str): _description_
        nf (str): _description_
        m (_type_): _description_
    """
    with open(nf, 'r') as i_f:
        all_text = i_f.read().strip().replace('\n', ' ').lower().split()
    for word in all_text:
        # print(word)
        lemmas = m.lemmatize(word)
        # doc = Doc(word)
        # print(doc.tokens)
        # print(doc.sents)
        # print(lemmas)
        # print()
        if lemmas[0] not in stop_words:
            sum_arr[key][lemmas[0]] +=1
        #break
    # print(sum_arr)
```

```python
NAME_START_POINT = '/home/al/Projects_My/hakaton_ozon_opg/dataset/'
list_dir = get_dir(NAME_START_POINT)
#print(list_dir)
array_txt: dict = {}
m = Mystem()
for sd in list_dir:
    key_one = sd.split('/')[-1]
    #print(key_one)
    array_txt[key_one] = Counter()
    list_dir_two = get_dir(sd)
    for dt in list_dir_two:
        list_text = glob(dt + '/*.txt')
        #print(list_text)
        #print()
        for nf in list_text:
            get_array_text(array_txt, key_one, nf, m)
            #break
        #break
    #break
set_dict = []
for key in array_txt:
    print()
    print(key)
    print(array_txt[key])
    dd = dict(array_txt[key])
    nab = []
    for kk in dd:
        if dd[kk] >=2 and kk not in stop_words:
            nab.append(kk)
    set_dict.append(nab)
    print()
```
